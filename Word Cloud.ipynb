```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import re
import string
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
```

```python
#read data mudik2021_artikel (terdapat 7 artikel didalamnya)
articles = open("/content/drive/MyDrive/Colab Notebooks/mudik2021_artikel.txt", "r")

data = articles.read()
```

```python
print ("Terdapat {} kata pada artikel".format(len(data)))
```

```python
frequency = {}
text_string = data.lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
#menghitung jumlah kata yang sering muncul
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
 
for words in frequency_list:
    print (words, frequency[words])
```

```python
#generate wordcloud
wordcloud = WordCloud(background_color='white', mode="RGB", width=1600, height=800).generate(data)

#plotting wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
#membuat list stopwords
stopwords = set(STOPWORDS)
stopwords.update(["yang", "dan", "di", "untuk", "atau", "dari", "ini"])

#generate wordcloud
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=1600, height=800).generate(data)

#plotting wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
#menampilkan value dari gambar kompas.jpeg
logo_kompas_mask = np.array(Image.open("/content/drive/MyDrive/Colab Notebooks/kompas.jpeg"))
logo_kompas_mask
```

```python
#generate wordcloud
wordcloud = WordCloud(background_color="white", mode="RGB", max_words=2000,  collocations=False, mask=logo_kompas_mask, stopwords=stopwords).generate(data)

#plotting wordcloud
plt.figure(figsize=(8,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
